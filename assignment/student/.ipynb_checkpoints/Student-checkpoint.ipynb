```python
#Importing the necessary library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the csv file (Datasets)
df = pd.read_csv('student-mat.csv', delimiter=';')
df.head(5)
# Display the column names and first few rows of the DataFrame to check the structure
print("Column Names:")
print(df.columns)

print("\nFirst few rows of the DataFrame:")
print(df.head())
```

```python
# Handle Missing Values and Data Types
print("\nMissing values in the DataFrame:")
print(df.isnull().sum())
```

```python
# List of unnecessary columns to drop
unnecessary_columns = ['school', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 
                       'traveltime', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'internet', 'romantic', 
                       'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health']

# Drop unnecessary columns if they exist in the DataFrame
df = df.drop(columns=[col for col in unnecessary_columns if col in df])
```

```python
print("DataFrame after dropping unnecessary columns:")
print(df.head())
```

```python
print("Column Names:")
print(df.columns)
```

```python
print("\nFirst few rows of the DataFrame:")
print(df.head())
```

```python
# Example: Bar plot of student ages
plt.figure(figsize=(10, 6))
sns.countplot(x='age', data=df)
plt.title('Distribution of Student Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of final grades (G3)
plt.figure(figsize=(10, 6))
sns.histplot(df['G3'], bins=20, kde=True)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar pie chart of student sex
sex_counts = df['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Student Sex')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix using only numeric columns
correlation_matrix = df[numeric_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

 Student Por

```python
df = pd.read_csv('student-por.csv', delimiter=';')
df.head(5)
```

```python
# Handle Missing Values and Data Types
print("\nMissing values in the DataFrame:")
print(df.isnull().sum())
```

```python
import numpy as np
# List of unnecessary columns to drop
unnecessary_columns = ['school', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 
                       'traveltime', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'internet', 'romantic', 
                       'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health']

# Drop unnecessary columns if they exist in the DataFrame
df = df.drop(columns=[col for col in unnecessary_columns if col in df])
# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)
# Drop rows with NaN values
df.dropna(inplace=True)
```

```python
# Ensure all numerical columns are in the correct type
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col])
        except ValueError:
            pass 
```

```python
# Display the first few rows of the DataFrame after dropping unnecessary columns
print("\nDataFrame after dropping unnecessary columns:")
print(df.head())
```

```python
 # Bar plot of student ages
plt.figure(figsize=(10, 6))
sns.countplot(x='age', data=df)
plt.title('Distribution of Student Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of final grades (G3)
plt.figure(figsize=(10, 6))
sns.histplot(df['G3'], bins=20, kde=True)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix using only numeric columns
correlation_matrix = df[numeric_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot of G1 vs G3
plt.figure(figsize=(10, 6))
sns.scatterplot(x='G1', y='G3', data=df)
plt.title('G1 vs G3')
plt.xlabel('G1')
plt.ylabel('G3')
plt.show()
```

```python
 # Pie chart of students' sex
sex_counts = df['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title('Distribution of Students by Sex')
plt.show()
```
