```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('CSV_for_EDA.csv', )

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Get basic information about the DataFrame
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Convert 'date' column to datetime type
df['date'] = pd.to_datetime(df['date'])
```

```python
# Ensure all columns have correct data types
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
# Remove outliers or incorrect data if any 
df = df[(df['inside_sales'] >= 0) & (df['outside_sales'] >= 0)]
```

```python
# Extract useful features from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek
```

```python
# Time Series Plots
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['inside_sales'], label='Inside Sales')
plt.plot(df['date'], df['outside_sales'], label='Outside Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Over Time')
plt.legend()
plt.show()
```

```python
# Distribution Plot
plt.figure(figsize=(12, 6))
sns.histplot(df['total_sales'], kde=True)
plt.title('Distribution of Total Sales')
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number]).columns
```

```python
# Compute the correlation matrix
corr_matrix = df[numeric_cols].corr()
```

```python
# Plot the correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='day_of_week', y='total_sales', data=df)
plt.title('Total Sales by Day of Week')
plt.show()
```
