```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('exam_score_prediction_data.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check data types
print(df.dtypes)
```

```python
# Summary statistics to identify outliers
print(df.describe())
```

```python
# Fill missing values with the mean of the column
df['StudyTime'].fillna(df['StudyTime'].mean(), inplace=True)
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)
df['PreviousScore'].fillna(df['PreviousScore'].mean(), inplace=True)
df['ExamScore'].fillna(df['ExamScore'].mean(), inplace=True)
```

```python
# Histogram for each column
df.hist(bins=20, figsize=(14, 10))
plt.suptitle('Histograms of Variables')
plt.show()
```

```python
# Pairplot to see relationships between variables
sns.pairplot(df)
plt.suptitle('Pairplot of Variables', y=1.02)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot of StudyTime vs. ExamScore
plt.figure(figsize=(8, 6))
sns.scatterplot(x='StudyTime', y='ExamScore', data=df)
plt.title('Study Time vs Exam Score')
plt.xlabel('Study Time')
plt.ylabel('Exam Score')
plt.show()
```

```python
# Scatter plot of Attendance vs. ExamScore
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Attendance', y='ExamScore', data=df)
plt.title('Attendance vs Exam Score')
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.show()
```

```python
# Box plot for identifying outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Variables')
plt.show()
```
