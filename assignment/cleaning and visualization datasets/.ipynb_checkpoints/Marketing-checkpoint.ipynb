```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('marketing_campaign.csv', sep='\t')

# Display basic info about the dataset
print(df.info())
```

```python
# Handle missing values (example: fill missing income with median)
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Distribution of Age (Year_Birth)
plt.figure(figsize=(10, 6))
sns.histplot(df['Year_Birth'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Year of Birth')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Income'], bins=30, kde=True)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot for numeric columns
sns.pairplot(df[['Income', 'Recency', 'NumDealsPurchases', 'NumWebPurchases', 'Response']], hue='Response')
plt.show()
```
