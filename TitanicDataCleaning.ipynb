```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
df = pd.read_csv("titanic.csv")
df.head()
```

```python
print("Initial Missing Values:\n", df.isnull().sum())
```

```python
print("Duplicated Values if present:", df.duplicated().sum())
```

```python
#Filling the missing values with the median age
df["Age"].fillna(df["Age"].median(), inplace = True)
```

```python
#Fill the missing Cabin values with placeholder "Unknown"
df["Cabin"].fillna("Unknown", inplace = True)
```

```python
#Filling the missing values with the most frequent values
most_frequent_embarked = df["Embarked"].mode()[0]
df["Embarked"].fillna(most_frequent_embarked, inplace = True)
```

```python
#Verifying the missing values have been handled
print("Cleaned Values:\n", df.isnull().sum())
```

```python
import seaborn as sns 
#Check for outliners in "Age" cloumn 
print("Age Statistics:")
print(df["Age"].describe())
```

```python
#Plot a box to visualize outliners in "Age"
plt.figure(figsize = (8,6))
sns.boxplot(x=df["Age"])
plt.title("Age Distribution with outliners")
plt.show()
```

```python
#Check for outliners in "Fare"
print("\nFare Statistics")
print(df["Fare"].describe())
```

```python
#Plot a box to visualize outliners in "Fare"
plt.figure(figsize = (8,6))
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution with outliners")
plt.show()
```

```python
#Import required libraries
from sklearn.preprocessing import LabelEncoder

#Create a label encoder onject
le = LabelEncoder()

#Encode "Sex" column
df["Sex"] = le.fit_transform(df["Sex"])

#Encode "Embarked" column
df["Embarked"] = le.fit_transform(df["Embarked"])
```

```python
df.head()
```

```python
#Extract the title from the "Name"
df["Title"] = df["Name"].str.extract("([A-Za-z]+)\.", expand=False)

#Replace rare titles with "Rare" and map other titles to a more concise representation
df["Title"] = df["Title"].replace(["Lady", "Capt", "Col", "Don", "Dr", "Major", "Rev", "Sir", "Dona"], "Rare")
df["Title"] = df["Title"].replace(["Mlle", "Ms"], "Miss")
df["Title"] = df["Title"].replace(["Mse"], "Mrs")
```

```python
df.head()
```

```python
#Create a new column "AgeGroup" based on age ranges
age_bins = [0, 16, 32, 48, 64, 80]
age_labels = ["Child", "Young Adult", "Adult", "Middle_Aged", "Senior"]
df["AgeGroup"] = pd.cut(df["Age"], bins = age_bins, labels = age_labels)
```

```python
df.head()
```

```python
#Create a new column "FareCategory" based on fare ranges
fare_bins = [0, 7.91, 14.454, 31, float("inf")]
fare_labels = ["Low", "Medium", "High", "Very High"]
df["FareCategory"] = pd.cut(df["Fare"], bins = fare_bins, labels = fare_labels)
```

```python
df.head()
```

```python
#Drop unnecessary columns
titanic_data = df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
titanic_data.head()
```

```python
#Calculate the survival rate by passenger class
survival_rate_by_class = titanic_data.groupby("Pclass")["Survived"].mean().reset_index()

#Create a bar blot
plt.figure(figsize = (8,6))
plt.bar(survival_rate_by_class["Pclass"], survival_rate_by_class["Survived"])
plt.xlabel("Passenger Class")
plt.title("Survival Rate by passenger Class")
plt.xticks(survival_rate_by_class["Pclass"], ["1st Class", "2nd Class", "3rd Class"])
plt.show()
```

```python
#Count the nimber of passenger by embarkarion part
embarkation_counts = titanic_data["Embarked"].value_counts()

#Create a pie char
plt.figure(figsize = (8,6))
plt.pie(embarkation_counts, labels = embarkation_counts.index, autopct = "%1.1f%%")
plt.axis("Equal")
plt.title("Proportion of passengers by Embarked Port")
plt.show()
```

```python
#Drop the unnecessary cloumns
columns_to_keep = ["Survived", "Pclass", "Sex", "SibSp", "Parch", "Fare"]
titanic_data = titanic_data(columns_to_keep)
titanic_data.head()
```

```python
import seaborn as sns

#Create a correlation matrix
cor_matrix = titanic_data.corr()

#Create a heatmap
plt.figure(figsize= (8,6))
sns.heatmap(cor_matrix, annot = True, cmap = "coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
